```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Read the data from the file
data = pd.read_csv('ICP.csv')
df=pd.DataFrame(data)
print(df)
```

```python
df2=df 
df = df.drop(df.index[-1])
print(df)
```

```python
df.columns = df.columns.str.replace(' ', '_')
print(df.columns)
```

```python
# Select the first row
first_row = df.iloc[0, 1:]

# Create a bar plot
first_row.plot(kind='bar', figsize=(10, 6), title='Bar plot of all columns for the first row')
plt.xlabel('Attributes')
plt.ylabel('Values')
plt.show()
# Create a combined bar plot for all rows
df.set_index('Sample_Id').T.plot(kind='bar', figsize=(14, 8), title='Combined Bar Plot for All Rows')
plt.xlabel('Attributes')
plt.ylabel('Values')
plt.legend(title='Sample Id', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
df['Sample_Id'] = df['Sample_Id'].str.replace(' d1000', '')
print(df['Sample_Id'])
```

```python
# Select the first row for the first 12 columns
first_row_12 = df2.iloc[6, 1:36]

# Create a bar plot for the first 12 columns with thinner bars
first_row_12.plot(kind='bar', figsize=(10, 6), title='ICP Concentrations of the Washing Water', color='green', width=0.1)
plt.ylabel('mg/L')
plt.show()

# Create a combined bar plot for the first 12 columns for all rows
df.iloc[:, :13].set_index('Sample_Id').T.plot(kind='bar', figsize=(14, 8), title='Concentration From ICP')
plt.ylabel('mg/L')
plt.legend(title='Sample Id', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# Create a combined bar plot for columns 10 to 23 for all rows
df.iloc[:, 0 and 10:24].set_index('Sample_Id').T.plot(kind='bar', figsize=(14, 8), title='Concentration From ICP (Columns 10 to 23)')
plt.ylabel('mg/L')
plt.legend(title='Sample Id', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
